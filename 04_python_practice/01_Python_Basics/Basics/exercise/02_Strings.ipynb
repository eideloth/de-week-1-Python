### Warm-ups

1. **Middle char(s)**

   * `middle(s)` → If odd length return the middle char, else return the two middle chars.

   ```python
   def middle(s):
       ...
   assert middle("abc") == "b"
   assert middle("abba") == "bb"
   ```

````python
def middle(s):
    return s[1:-1]
    
assert middle("abc") == "b"
assert middle("abba") == "bb"
````


2. **Title-case safely**

   * `safe_title(s)` → title case but do not lowercase words fully in ALL CAPS acronyms (e.g., `"learn SQL now"` → `"Learn SQL Now"`; `"use GPU"` → `"Use GPU"`).

   ```python
   def safe_title(s):
       ...
   ```

````python
def safe_title(s):
    title = s.split();
    title2 = []
    for e in title:
        if(e[0].islower()):
            title2.append(e.capitalize())
        else:
            title2.append(e)    
    return " ".join(title2)
    
assert safe_title("learn SQL now") == "Learn SQL Now"
````


3. **Reverse words**

   * `reverse_words(s)` → reverse word order but keep internal characters.

   ```python
   def reverse_words(s):
       ...
   assert reverse_words("one two  three") == "three two  one"
   ```


````python
def reverse_words(s):
    ws = s.count(" ")
    words = s.split()[::-1]
    return ws.join(words)
#reverse_words("one two  three")
assert reverse_words("one two  three") == "three two  one"
````

### Core

4. **Normalize spaces**

   * `squash_spaces(s)` → collapse multiple spaces/tabs to a single space; strip ends.

   ```python
   def squash_spaces(s):
       ...
   assert squash_spaces("  a\t b   c ") == "a b c"
   ```


````python
def squash_spaces(s):
    i = s.split()
    result = " ".join(i)
    return result
    
assert squash_spaces("  a\t b   c ") == "a b c"
````

5. **CSV line → list (no quotes)**

   * `split_csv(s)` → split by commas, trim whitespace around fields; ignore empty trailing field if line ends with comma.

   ```python
   def split_csv(s):
       ...
   assert split_csv(" a, b ,c,") == ["a","b","c"]
   ```

````python
def split_csv(s):
    return s.replace(",", " ").split()   


assert split_csv(" a, b ,c,") == ["a","b","c"]
````

6. **Mask secrets**

   * `mask_email(s)` → keep first char of user and domain, mask the rest with `*`, keep TLD.

     * `"alice@example.com"` → `"a****@e******.com"`

   ```python
   def mask_email(s):
       ...
   ```

````python
def mask_email(s):
    user, domain = s.split("@", 1)
    name, tld = domain.rsplit(".", 1)
    masked_user = user[0] + "*" * (len(user) - 1) if user else ""
    masked_domain = name[0] + "*" * (len(name) - 1) if name else ""

    return f"{masked_user}@{masked_domain}.{tld}"

assert mask_email("alice@example.com") == "a****@e******.com"
````

7. **Find all indexes**

   * `find_all(s, sub)` → list of start indices where `sub` occurs (including overlaps).

   ```python
   def find_all(s, sub):
       ...
   assert find_all("aaaa", "aa") == [0,1,2]
   ```


````python
def find_all(s, sub):
    findings = []
    length = len(s)
    index = 0
    while index < length:
        find = s.find(sub, index)
        if(find >= 0):
            findings.append(s.find(sub, index))
        index = index + 1
    return findings
    

assert find_all("aaaa", "aa") == [0,1,2]
````

8. **Anagram check**

   * `is_anagram(a, b)` ignoring spaces, case, punctuation.[Anagram](https://en.wikipedia.org/wiki/Anagram)

   ```python
   def is_anagram(a, b):
       ...
   assert is_anagram("Listen", "Silent")
   ```

````python
def is_anagram(a, b):
    return sorted(a.lower()) == sorted(b.lower()) 
    
assert is_anagram("Listen", "Silent")
````
